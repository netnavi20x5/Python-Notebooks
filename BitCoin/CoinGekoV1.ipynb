```python
cg = CoinGeckoAPI()
time.sleep(3)
coins =['bitcoin', 'bitcoin-cash', 'ethereum', 'litecoin', 'ripple']
coin_id=cg.get_coins_list()
for coin in coins:
    print (coin)
    coin_market[coin]=(cg.get_coin_market_chart_by_id(id=coin,vs_currency="MYR",days="90",interval="daily"))
```

```python
##Functions
def convert_unix(time): #For fuck sake this is annoying To convert time 
    return datetime.datetime.fromtimestamp(int(str(time)[:10])).strftime('%Y-%m-%d %H:%M:%S')
    #return datetime.datetime.fromtimestamp(int(str(time)[:10])).strftime('%Y-%m-%d')

def convert_date_list(days,day): #annoying function 
    reverse=list()
    for i in day:
        reverse.append(days-i)
    return reverse

def check_price_diff_days(dict_data,coin):
    data =dict_data
    days=len(dict_data['prices'])
    for i in range (days):
        percent=((data["prices"][days-1][1]/data["prices"][i][1]-1)*100)
        print (f"{str(days-i-1)} days ago difference for {coin} by {str('%.2f' % percent)} ")

def check_price_diff_week_14_20_30(dict_data,coin):
    days=len(dict_data['prices'])
    data=dict_data
    output=list()
    for i in range (days):
        if i in convert_date_list(days,[1,2,3,4,5,6,7,14,10,20,30]):
            percent=((data["prices"][days-1][1]/data["prices"][i][1]-1)*100)
            print (f"{str(days-i-1)} days ago difference for {coin} by {str('%.2f' % percent)} ")
    return output

def check_price_diff_daily_days(dict_data,coin):
    days=len(dict_data['prices'])
    data=dict_data
    output=list()
    for i in range (1,days):
        calc=data["prices"][i][1]/data["prices"][i-1][1]-1
        print(f"Difference between {convert_unix(data['prices'][i][0])} and {convert_unix(data['prices'][i-1][0])} for {coin} is : {calc*100}")


def check_price_diff_days(dict_data,coin,limit):
    days=len(dict_data['prices'])
    data=dict_data
    for i in range (days-limit-1,days):
        percent=((data["prices"][days-1][1]/data["prices"][i][1]-1)*100)
        print (f"{str(days-i-1)} days ago difference for {coin} by {str('%.2f' % percent)} ")

def check_price_diff_daily_days(dict_data,coin,limit):
    days=len(dict_data['prices'])
    data=dict_data
    for i in range (days-limit-1,days):
        calc=data["prices"][i][1]/data["prices"][i-1][1]-1
        print(f"Difference between {convert_unix(data['prices'][i][0])} and {convert_unix(data['prices'][i-1][0])} for {coin} is : {calc*100}")
```

```python
for coin in coins:
    print (coin)
    print ("Difference in a day")
    check_price_diff_daily_days(coin_market[coin],coin,5)
    print ("Difference in a week, 10 days, 2 weeks, 20 days and 30 days ")
    check_price_diff_week_14_20_30(coin_market[coin],coin)
    print ("One week")
    check_price_diff_days(coin_market[coin],coin,5)
```

```python
def convert_data_frame(dict_data):
    df =pd.DataFrame(dict_data["prices"])
    df.columns = ['posix', 'price']
    #df.date=df.apply(lambda row : add(row['A'],row['B'], row['C']), axis = 1)vconvert_unix(time)
    df["date"]=df.apply(lambda row : convert_unix(row['posix']), axis = 1)
    price_now=df['price'].values[len(df.price)-1]
    df["diff_between_now"]=(price_now/df['price']-1)*100
    df["diff_previous_day"]= df.price.pct_change() * 100
    return df[["date","price","diff_between_now","diff_previous_day"]]
convert_data_frame(coin_market['bitcoin'])
```
