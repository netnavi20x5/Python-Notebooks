```python
url="https://wnarifin.github.io/covid-19-malaysia/#overview"
```

```python
import requests
from bs4 import BeautifulSoup
x = requests.get(url)
soup = BeautifulSoup(x.text, 'html.parser')
```

```python
overview=soup.find(id="overview")
```

```python
for i in overview.find_all("div"):
    print ("===============================================")
    print (i)
    
```

```python
row=overview.find_all("div",{"class":"section level2"})
for i,item in  enumerate (row):
    print (item.find_all("span"))
```
