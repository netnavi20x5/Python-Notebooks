```python
import numpy
import pandas as pd
```

```python
registration="https://raw.githubusercontent.com/CITF-Malaysia/citf-public/main/registration/"
static_citf="https://raw.githubusercontent.com/CITF-Malaysia/citf-public/main/static/"
vaccination="https://raw.githubusercontent.com/CITF-Malaysia/citf-public/main/vaccination/"


#registration
df_vaxreg_malaysia=pd.read_csv(registration+"vaxreg_malaysia.csv")
df_vaxreg_state=pd.read_csv(registration+"vaxreg_state.csv")
#vaccination
df_vax_malaysia=pd.read_csv(vaccination+"vax_malaysia.csv")
df_vax_state=pd.read_csv(vaccination+"vax_state.csv")

df_population_citf=pd.read_csv(static_citf+"population.csv")

```

```python
latest_date=df_vaxreg_state.tail(1)["date"].values[0]
df_latest_vaxreg_state=df_vaxreg_state[df_vaxreg_state.date==latest_date]
df_latest_vax_state=df_vax_state[df_vax_state.date==latest_date]

#df_latest_vax_state.join(df_latest_vaxreg_state.set_index('date'), onstatestate')

df_vaxreg_x_vax=df_latest_vaxreg_state.join(df_latest_vax_state.set_index('state'), on='state',lsuffix='_vaxreg', rsuffix='_vax')[['state', 'total', 'dose1_cumul', 'dose2_cumul',
       'total_cumul']]
df_vaxreg_x_vax["dose1_percent"]=df_vaxreg_x_vax["dose1_cumul"]/df_vaxreg_x_vax["total"]*100
df_vaxreg_x_vax["dose2_percent"]=df_vaxreg_x_vax["dose2_cumul"]/df_vaxreg_x_vax["total"]*100
df_vaxreg_x_vax.sort_values(by=['dose2_percent'],ascending=False)
```
