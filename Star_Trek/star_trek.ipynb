```python
def extract_lines(series,cast):
    df_all=pd.read_json("data/all_series_lines.json")
    lines=""
    crew=list()
    df_series=pd.DataFrame(df_all[series]).dropna()
    for i in df_series.values:
        try:
            crew.append(i[0][cast])
        except:
            print("An exception occurred")
    print("successful ")
    for lvl1 in crew:
        for lvl2 in lvl1:
            #print(lvl2)
            lines=lines+lvl2
    return lines
#df_tos=pd.DataFrame(df_all["TOS"].dropna())
#df_tos.reindex(range(1,175))

lines=extract_lines("TOS","SPOCK")
#lines.replace("\\",' ')
```

```python
stopwords=set(STOPWORDS)
#stopwords.add("")
wcloud=WordCloud(background_color='white',colormap="plasma",stopwords=stopwords).wcloud=WordCloud().generate(lines)
plt.imshow(wcloud)
```

```python
spock=np.array(Image.open("spock.jpg"))
spock_wc=WordCloud(background_color='white',colormap="plasma",stopwords=stopwords,max_words=2000,max_font_size=2000, mask = spock)
spock_wc.generate(lines)
image_colors=ImageColorGenerator(spock)

fig,axes = plt.subplots(1,2)
axes[0].imshow(spock,interpolation="bilinear", cmap = plt.cm.gray)
axes[1].imshow(spock_wc.recolor(color_func=image_colors),interpolation="bilinear")

for ax in axes:
    ax.set_axis_off()
plt.show()


plt.imshow(spock_wc.recolor(color_func=image_colors),interpolation="bilinear")
plt.axis("off")
plt.show()
```
